```python
import pandas as pd
import gc
gc.collect()
```

```python
#打开文件
f = open('result.csv', 'r')
df = pd.DataFrame(columns=['title', 'price', 'purchace_num', 'url'])
```

```python
#读取文件
lines = f.readlines()
for line in lines:
    kv = line.split(',')
    df = df.append(pd.DataFrame({'title': [kv[0]], 'price': [kv[1]], 'purchace_num': [kv[2]], 'url': [kv[3]]}))

df = df.reset_index(drop=True)
```

```python
df.head()
```

```python
#数据处理
for i in range(len(df)):
    df.ix[i, 'purchace_num'] = df.ix[i, 'purchace_num'][:-3]  #去除 人付款
    if '+' in df.ix[i, 'purchace_num'] :
        df.ix[i, 'purchace_num'] = df.ix[i, 'purchace_num'][:-1] #去除+号
    if '万' in df.ix[i, 'purchace_num'] :
        df.ix[i, 'purchace_num'] = df.ix[i, 'purchace_num'][:-1]  #去除 万
    if '.' in df.ix[i, 'purchace_num'] :
        int_, digit_ = df.ix[i, 'purchace_num'].split('.')
        df.ix[i, 'purchace_num'] = int_ + digit_ + '000'   #'改变.0结构'
```

```python
index_drop = []
for i in range(len(df)):
    try:
        df.ix[i, 'purchace_num'] = int(df.ix[i, 'purchace_num'])  #int
    except:
        index_drop.append(i)
```

```python
df = df.drop(index=index_drop)
df = df.reset_index(drop=True)
```

```python
for i in range(len(df)):
    df.ix[i, 'price'] = df.ix[i, 'price'][1:]
    
```

```python
df.head()
```

```python
#选择大于销量>100的行
df_ = df[df.purchace_num > 100]
```

```python
df_ = df_.sort_values(by='purchace_num', ascending=False).reset_index(drop=True)
```

```python
df_.head()
```

```python
for i in range(len(df_)):
    df_.iloc[i, 3] = df_.iloc[i, 3][2: -2]
```

```python
df_.to_csv('n95_>100.csv', index=False)
```

```python
df_o = pd.read_csv('out.csv', encoding='GB18030', header=0, names=['date', 'price', 'title', 'date2'])
del df_o['date2']
df_o=df_o.drop_duplicates(subset=None,keep='first',inplace=False) #去除重复行
df_o = df_o.dropna(how='any')  #去除空值
df_o = df_o.reset_index(drop=True)
```

```python
df_o
```

```python
#涨幅计算
for i in range(len(df_o)):
    if df_o.iloc[i, 0] == '2020-01-20':
        df_o.ix[i, '20up'] = (df_o.iloc[i, 1] - df_o.iloc[i-1, 1]) / df_o.iloc[i-1, 1]
    if df_o.iloc[i, 0] == '2020-01-21':
        df_o.ix[i, '21up'] = (df_o.iloc[i, 1] - df_o.iloc[i-1, 1]) / df_o.iloc[i-1, 1]
    if df_o.iloc[i, 0] == '2020-01-22':
        df_o.ix[i, '22up'] = (df_o.iloc[i, 1] - df_o.iloc[i-1, 1]) / df_o.iloc[i-1, 1]
    
```

```python
df_o
```

```python
import datetime
df_o['date_'] = pd.to_datetime(df_o['date'])
time_ = datetime.datetime(2020, 1, 19)
mask = df_o['date_'] > time_
data = df_o[mask]
```

```python
data_20 = data[['date', '20up']].dropna(how='any').sort_values(by='20up', ascending=False).reset_index(drop=True)
data_21 = data[['date', '21up']].dropna(how='any').sort_values(by='21up', ascending=False).reset_index(drop=True)
data_22 = data[['date', '22up']].dropna(how='any').sort_values(by='22up', ascending=False).reset_index(drop=True)

```

```python
data_20.to_csv('20.csv')
data_21.to_csv('21.csv')
data_22.to_csv('22.csv')
```

```python
data_20_up = len(data_20[data_20['20up']>0]) / len(data_20) * 100
data_21_up = len(data_21[data_21['21up']>0]) / len(data_21) * 100
data_22_up = len(data_22[data_22['22up']>0]) / len(data_22) * 100
print(data_20_up, data_21_up, data_22_up)
```
